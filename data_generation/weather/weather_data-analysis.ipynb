We are interested in:
- Temperature 
- Precipitation
- Wind

date range we are looking: 2019-01-02 to 2019-06-30.

- Download them in https://opendata.dwd.de/climate_environment/CDC/observations_germany/climate/10_minutes/  -> get folders of weather/raw

NOTE: QN - quality . 3 means Fully quality-controlled and corrected.

**Precipitation**

```python
import os
import pandas as pd
from IPython.display import display

DATA_ROOT = r"C:\Users\user\UPM\Imperial-4año\IoT\Github\hugging_face\weather_berlin-tempel"

# ---- Load CLEANED temperature metadata ----
meta_path = os.path.join(DATA_ROOT, "interest", "metadata", "precipitation.csv")
meta_raw = pd.read_csv(meta_path)

display(meta_raw)
```

```python
# only var names etc
meta = meta_raw[(meta_raw["From_Date"] <= 20190102) & (meta_raw["To_Date"] >= 20190630)]
meta = meta[["Parameter", "Parameter_Description", "Unit"]]
display(meta)

# ---- Load raw precipitation data ----
raw_path = os.path.join(
    DATA_ROOT,
    "raw",
    "precipitation",
    "10minutenwerte_nieder_00433_20100101_20191231_hist",
    "produkt_zehn_min_rr_20100101_20191231_00433.txt",
)

df = pd.read_csv(raw_path, sep=";", engine="python")

# Remove eor column if present
df = df.loc[:, ~df.columns.str.contains("eor")]

# Convert date
df["MESS_DATUM"] = pd.to_datetime(df["MESS_DATUM"], format="%Y%m%d%H%M")

# Filter
mask = (df["MESS_DATUM"] >= "2019-01-02") & (df["MESS_DATUM"] <= "2019-06-30")
df = df.loc[mask]

print("\n=== Filtered precipitation data ===")
display(df)

# ---- Save output ----
save_path = os.path.join(DATA_ROOT, "interest", "precipitation.csv")
df.to_csv(save_path, index=False)

print(f"\nSaved filtered precipitation data -> {save_path}")
```

```python
num_cols = [c for c in df.select_dtypes(include="number").columns 
            if c not in ["STATIONS_ID", "MESS_DATUM"]]

print(df[num_cols].describe().T[["mean", "std", "min", "max"]])
```
